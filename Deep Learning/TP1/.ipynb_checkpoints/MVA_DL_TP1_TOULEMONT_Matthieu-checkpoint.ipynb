```python
from sklearn.preprocessing import StandardScaler

save_to = "cnn_reg"
mcp_reg = makeMCP(save_to)
```

```python
def sort_by_x(Y):
    Z = np.zeros_like(Y)
    for row in range(Y.shape[0]): 
        y = Y[row].reshape(3,2)
        Z[row] = y[y[:,0].argsort(),:].flatten()
    return Z
```

```python
normalizer = StandardScaler()
y_train_normalize = normalizer.fit_transform(Y_train_reg)
y_test_normalize = normalizer.transform(Y_test_reg)

Xr_train = X_train.reshape(5000,72,72,1)
Xr_test = X_test.reshape(300,72,72,1)
visualize_prediction(X_train[0], Y_train_reg[0])
```

```python
def cnn_reg():
    model = Sequential()
    model.add(Conv2D(40, kernel_size=(5,5), strides = 1, activation = 'relu'))
    model.add(BatchNormalization())
    model.add(MaxPool2D(pool_size = (2,2), strides = 2))
    model.add(Conv2D(20, kernel_size=(4,4), strides = 1, activation = 'relu'))
    model.add(BatchNormalization())
    model.add(MaxPool2D(pool_size = (2,2), strides = 2))
    model.add(Flatten())
    model.add(Dense(6, activation = None))
    model.compile(loss="mse", optimizer="adam")
    return model
```

```python
model = cnn_reg()

history = model.fit(Xr_train, y_train_normalize,validation_split=0.1, epochs=40, batch_size=500, verbose = 1, callbacks = [mcp])
print(model.evaluate(Xr_train, y_train_normalize))
print(model.evaluate(Xr_test, y_test_normalize))
```
